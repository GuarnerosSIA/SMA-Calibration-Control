# Super-Twisting Algorithm

The STA application as an robust estimator derivator is described as follows. If $w_1(t) = r(t)$ where $r(t)\in \mathbb{R}$  is the signal to bhe differentiated, $w_2(t)=\dot{r}(t)$ represents its derivative and unde rthe assumption of $\Vert \ddot{r} \Vert \leq r^{+}$. the following differential equation is got CAMBI

\begin{equation}
\begin{array}{c}
\dot{w}_1(t) = w_2(t) \\
\dot{w}_2(t) = \ddot{r}(t) \\
\end{array}
\end{equation}

The previous differential equation is a state representation of the signal $r(t)$. The scalar STA to nobatin the the derivative of $r(t)$ looks like

\begin{equation}
\begin{array}{c}
\displaystyle\frac{d}{dt}\bar{W}_1(t) = \bar{W}_2(t) - \lambda_1 \Vert e_1(t) \Vert^{1/2}sign(e_1(t)) \\
\displaystyle\frac{d}{dt}\bar{W}_2(t) = - \lambda_2 sign(e_1(t))
\end{array}
\end{equation}
with $e_1(t) = \bar{W}_1-W_1(t)$

```
import numpy as np
import matplotlib.pyplot as plt
```

```
tau = 0.01
T = int(10/0.01)
x = np.linspace(0,10,T)
r = np.sin(x)
dr = np.cos(x)
W1 = np.zeros((T))
W2 = np.zeros((T))
plt.plot(x,r)
plt.plot(x,dr)
plt.plot(x,W1)
plt.plot(x,W2)
```

```
l1 = 2
l2 = 3
for i in range(T-1):
    e = W1[i]-r[i]
    W1[i+1] = W1[i]+tau*(W2[i] - l1*(np.sqrt(np.abs(e))*np.sign(e)))
    W2[i+1] = W2[i]+tau*(-l2*np.sign(e))
    
plt.plot(x,dr)
plt.plot(x,W2)
```

```
plt.plot(x,W1)
plt.plot(x,W2)
```
